# Cleaning The Data

```python
# Most of the years have non year values
car=car[car['year'].str.isnumeric()]
```

```python
car
```

```python
# Changing the object data type of year to integer
car["year"]=car["year"].astype(int)
```

```python
car
```

```python
car.info()
```

```python
# Taking those price values which have a value not this ("Ask For Price")
car=car[car["Price"] != "Ask For Price"]
```

```python
car.shape
```

```python
# price have some comma values
car["Price"]=car["Price"].str.replace(",","").astype(int)
```

```python
car
```

```python
car.info()
```

```python
def updateKms_drive(text):
    y=text.split()[0]
    p="".join(y)
    return p.replace(",","")
```

```python
updateKms_drive("45,000 kms")
```

```python
car["kms_driven"]=car["kms_driven"].apply(updateKms_drive).astype(int)
```

```python
car.info()
```

```python
car.head()
```

```python
def updateName(text):
    y=text.split()
    return " ".join(y[0:3])
```

```python
updateName("Hyundai Santro Xing XO eRLX Euro III")
```

```python
# Changing the car names and only considering the first three word
car["name"]=car["name"].apply(updateName)
```

```python
x=car[["name","company","year","kms_driven","fuel_type"]]
```

```python
x
```

```python
y=car["Price"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.1,random_state=302)
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
lr=LinearRegression()
ohe=OneHotEncoder()
ohe.fit(x[["name","company","fuel_type"]])
```

```python
ohe.categories_
```

```python
from sklearn.compose import make_column_transformer
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),["name","company","fuel_type"]),remainder="passthrough")
```

```python
column_trans
```

```python
from sklearn.pipeline import make_pipeline
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(xtrain,ytrain)
```

```python
ypred=pipe.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,ypred)
```

```python
scores=[]
for i in range(1000):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(xtrain,ytrain)
    ypred=pipe.predict(xtest)
    scores.append(r2_score(ytest,ypred))
```

```python
np.argmax(scores)
```

```python
max(scores)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(criterion="friedman_mse",max_depth=1007)
```

```python
pipe=make_pipeline(column_trans,dt)
```

```python
pipe.fit(xtrain,ytrain)
ypred=pipe.predict(xtest)
r2_score(ytest,ypred)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(criterion="friedman_mse",n_estimators=70)
pipe=make_pipeline(column_trans,rfr)
pipe.fit(xtrain,ytrain)
ypred=pipe.predict(xtest)
r2_score(ytest,ypred)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor(learning_rate=0.1,n_estimators=4000)
pipe=make_pipeline(column_trans,gbr)
pipe.fit(xtrain,ytrain)
ypred=pipe.predict(xtest)
r2_score(ytest,ypred)
```
